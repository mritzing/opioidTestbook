```python
%matplotlib inline
import pandas as pd
import os
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
sns.set(style="white", rc={"axes.facecolor": (0, 0, 0, 0)})

#read CSV to pandas dataframe http://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.DataFrame.html
#print shape (length/width) and sample points to see it was read in correctl 
df = pd.read_csv('Sheet2 (All Opioids - Age Group and Year)_clean.csv',header= 0)
print (df.shape)
print (df['Year'].unique())
print (df.sample(3))
```

#Heatmap in Seaborn
https://seaborn.pydata.org/examples/heatmap_annotation.html


```python
dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=dims)
# Need to make year group into categories to preserve order https://stackoverflow.com/questions/18189981/reordering-columns-rows-of-a-pivot-table
df["Ten-Year Age Groups"] = df["Ten-Year Age Groups"].astype('category', categories=['< 1 year','1-4 years','5-14 years','15-24 years','25-34 years','35-44 years','45-54 years','55-64 years','65-74 years','75-84 years','85+ years','Not Stated'])
result = df.pivot(index='Year', columns='Ten-Year Age Groups',values='Crude Rate')
ax = sns.heatmap(result, annot=True,fmt="g", cmap='viridis')
ax.set_yticklabels(ax.get_yticklabels(),rotation=0)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
ax.invert_yaxis()
```
